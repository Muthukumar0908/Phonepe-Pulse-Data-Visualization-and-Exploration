```python
path5=r"E://phonepee//pulse-master//pulse-master//data//top//transaction//country//india//state//"
top_transaction_list=os.listdir(path5)
# print(agg_state_list)
columns_name={'State':[], 'Year':[],'Quater':[],'district':[], 'type':[], 'count':[],'amount':[]}
for i in top_transaction_list:
    p_i=path5+i+"//"
    # print(p_i)
    year1=os.listdir(p_i)
    # print(Agg_yr)
    for j in year1:
        p_j=p_i+j+"//"
        agg_yr_list=os.listdir(p_j)
        for k in agg_yr_list:
            p_k=p_j+k 
            Data=open(p_k,'r')
            D=json.load(Data)
            D
            for z in D['data']['districts']:
              Name=z['entityName']
              type=z['metric']['type']
              count=z['metric']['count']
              amount=z['metric']['amount']
              columns_name['district'].append(Name)
              columns_name['type'].append(type)
              columns_name['count'].append(count)
              columns_name['amount'].append(amount)
              columns_name['State'].append(i)
              columns_name['Year'].append(j)
              columns_name['Quater'].append(int(k.strip('.json')))
df_top_transaction=pd.DataFrame(columns_name)
```

```python
path6=r"E://phonepee//pulse-master//pulse-master//data//top//user//country//india//state//"
top_user_list=os.listdir(path6)
columns_name={'State':[], 'Year':[],'Quater':[],'district_name':[], 'register_dist':[], 'pincode':[],'registeredUsers':[]}
for i in top_user_list:
    p_i=path6+i+"//"
    # print(p_i)
    year1=os.listdir(p_i)
    for j in year1:
        p_j=p_i+j+"//"
        agg_yr_list=os.listdir(p_j)
        for k in agg_yr_list:
            p_k=p_j+k 
            Data=open(p_k,'r')
            D=json.load(Data)
            # D
            for z in D['data']['districts']:
              district=z['name']
              registeredUsers=z['registeredUsers']

              
            for y in D['data']['pincodes']:
              pincode=y['name']
              registeredUsers=y['registeredUsers']
              columns_name['district_name'].append(district)
              columns_name['register_dist'].append(registeredUsers)
              columns_name['pincode'].append(pincode)
              columns_name['registeredUsers'].append(registeredUsers)
              columns_name['State'].append(i)
              columns_name['Year'].append(j)
              columns_name['Quater'].append(int(k.strip('.json')))
df_top_user=pd.DataFrame(columns_name)
```
